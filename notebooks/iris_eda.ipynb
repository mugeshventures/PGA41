```python
# Imports: standard data analysis and visualization libraries
import pandas as pd  # data handling
import numpy as np  # numeric ops
import matplotlib.pyplot as plt  # plotting
import seaborn as sns  # statistical plotting
# Set plotting style
sns.set(style="whitegrid")
%matplotlib inline  # For notebook inline plots (commented if running outside Jupyter)
```

```python
# Load the dataset from the workspace path
# Use raw string because the filename contains a space
data_path = r'data/Iris dataset.csv'
# Read CSV into a DataFrame
df = pd.read_csv(data_path)
# Show the shape and first rows to validate load
df.shape  # (rows, columns)
df.head()
```

```python
# Basic info and statistics
# Print column dtypes and non-null counts
df.info()
# Summary statistics for numeric columns
df.describe().T
```

```python
# Check for missing values and duplicates
# Missing values per column
df.isnull().sum()
# Count duplicate rows
df.duplicated().sum()
```

```python
# Rename columns to clean names (optional)
# We'll create a copy to avoid accidental modification of the original immediately
data = df.copy()
# Drop Id column if present as it's just an index
if 'Id' in data.columns:
    data = data.drop(columns=['Id'])
# Standardize column names to snake_case for convenience
data.columns = [c.strip().replace(' ', '_').replace('Cm','cm').lower() for c in data.columns]
data.head()
```

```python
# Univariate analysis: distributions for numeric features
# Histogram and KDE for each numeric column
numeric_cols = data.select_dtypes(include=[np.number]).columns.tolist()
# Plot histograms in a grid
plt.figure(figsize=(12, 8))
for i, col in enumerate(numeric_cols, 1):
    plt.subplot(2, 2, i)
    sns.histplot(data[col], kde=True, bins=20)
    plt.title(f'Distribution of {col}')
plt.tight_layout()
```

```python
# Target (species) distribution
# The original CSV uses 'Species' with values like 'Iris-setosa'
if 'species' in data.columns:
    plt.figure(figsize=(6,4))
    sns.countplot(x='species', data=data)
    plt.title('Species distribution')
    plt.xticks(rotation=45)
else:
    print('Column 
 not found; check header names')
```

```python
# Pairplot to inspect relationships between features colored by species
# This can be expensive for very large datasets but is fine for Iris (~150 rows)
if 'species' in data.columns:
    sns.pairplot(data, hue='species', corner=True)
else:
    print('Cannot create pairplot: missing target column')
```

```python
# Correlation matrix and heatmap
corr = data.select_dtypes(include=[np.number]).corr()
plt.figure(figsize=(6,5))
sns.heatmap(corr, annot=True, cmap='coolwarm', fmt='.2f')
plt.title('Feature Correlation')
```

```python
# Boxplots per species to spot differences and outliers
if 'species' in data.columns:
    plt.figure(figsize=(12,8))
    for i, col in enumerate(numeric_cols, 1):
        plt.subplot(2, 2, i)
        sns.boxplot(x='species', y=col, data=data)
        plt.title(f'Boxplot of {col} by species')
    plt.tight_layout()
else:
    print('Boxplots skipped: target column missing')
```
